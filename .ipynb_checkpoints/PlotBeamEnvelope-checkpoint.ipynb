```python
from numpy import sqrt
import numpy as np
import matplotlib.pyplot as plt
from itertools import permutations
from scipy.signal import argrelextrema

#auxiliary function definitions
def F(f):
    mat = np.matrix([[1,0],[-1/f,1]])
    return mat

def O(d):
    mat = np.matrix([[1,d],[0,1]])
    return mat

def q_prop(q,mat):
    A=mat[0,0]
    B=mat[0,1]
    C=mat[1,0]
    D=mat[1,1]
    qout=(A*q+B)/(C*q+D)
    return qout

def q_backprop(q,mat):
    A=mat[0,0]
    B=mat[0,1]
    C=mat[1,0]
    D=mat[1,1]
    qout = (B-D*q)/(C*q-A)
    return qout

def q(w0,d,lam):   
    qout = d+1j*3.14159*w0*w0/lam
    return qout

def w(q,lam):
    thing=1/q
    thing2=-thing.imag
    wout=sqrt(lam/(3.14159*thing2))
    return wout

def w0(q,lam):
    wout= sqrt(lam*q.imag/3.14159)
    return wout

def decompose(mat):
    A = mat[0,0]
    B = mat[0,1]
    C = mat[1,0]
    D = mat[1,1]
    return ([A,B,C,D])
```

```python
#the spatial distance the beam will travel
start = 0;
stop = 0.44661;
sampling = 1000;
spacing = (stop-start)/sampling;
```

```python
#Beam parameters
z0 = 0;
zr0 = 1;
lam0 = 1550e-9;
q0 = z0 + zr0*1j;
```

```python
#the telescope, lens by lens (alternate distance and lens)
#for example, here I have a 60 mm drift, followed by a 150 mm lens, 
#followed by a 71.27 mm drift, followed by a -75 mm lens, etc

dArray = ([60e-3,71.27e-3,210.34e-3])
dRemain = stop-sum(dArray)
dArray = np.append(dArray,dRemain)
lArray = ([150e-3,-75e-3,75e-3])

#intialize array
Mt = O(0)

#initialize q
qCurr = q0;
zCurr = start;
q = ([])
z = ([])


#iterate through array and construct transfer matrix
for elem in zip(dArray,lArray):
    
    #distance to propagate
    zIter = np.arange(0,elem[0],spacing)
    #generate vector of qs over some distance
    for zVal in zIter:
        q = np.append(q,qCurr + zVal)
        z = np.append(z,zCurr + zVal)
    
    #update current z value
    zCurr = z[-1]
    #update current q value
    qCurr = q[-1]
    
    #transform q with lens
    c = -1/elem[1]
    qCurr = qCurr/(1 + c * qCurr)
    
    #transfer matrix update
    Mt = F(elem[1])*O(elem[0])*Mt
    
#propagate through remaining distance
zRem = np.arange(0,dArray[-1],spacing)
for zVal in zRem:
    q = np.append(q,qCurr + zVal)
    z = np.append(z,zCurr + zVal)

#transfer matrix update
Mt = O(dArray[-1])*Mt
    
#update current z value
zCurr = z[-1]
#update current q value
qCurr = q[-1]

#find w from q
wArray = w(q,lam0)
```

```python
#plot them
plt.plot(z*1e3,wArray*2e3)
plt.xlabel("distance (mm)")
plt.ylabel("beam diameter (mm)")
plt.show()
print("=======================")
print("Final Transfer Matrix:")
print(Mt)
print("=======================")
print("Focusing?")
print(abs(Mt[0,0]) < 1e-5)
print("=======================")
print("Imaging?")
print(abs(Mt[0,1]) < 1e-5)
print("=======================")
print("Spatial Magnification:")
print(Mt[0,0])
print("=======================")
print("Angular Magnification:")
print(Mt[1,1])
print("=======================")
print("Effective Focal Length (mm):")
print(-1e3/Mt[1,0])
```
